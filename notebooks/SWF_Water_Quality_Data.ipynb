### Water quality data retrieved from https://insightsdata.sfwmd.gov/#/waterquality

```python
import pandas as pd
```

```python
wq_data = pd.read_csv(r'C:\Users\ianra\Downloads\sfwmd-data-202603191442.csv', comment='#', low_memory = False)
wq_data.head()
```

```python
print(wq_data.columns.tolist())
print(wq_data.dtypes)
print(wq_data.shape)
#list of all column names, stations, data types, dataframe shape
```

```python
wq_data['collectDate'] = pd.to_datetime(wq_data['collectDate'])
wq_data = wq_data.set_index('collectDate')
wq_data.head()
```

```python
print(wq_data.index.min()) 
print(wq_data.index.max())  

```

```python
# Check which columns you actually need
print(wq_data.columns.tolist())

# Check how many unique stations you have
print(f"Unique stations: {wq_data['station'].nunique()}")
print(wq_data['station'].unique())

# Check which parameters are available
print(f"Unique parameters: {wq_data['parameter'].nunique()}")
print(wq_data['parameter'].unique())
```

```python
col_to_keep = [
    'station', 'project', 'parameter', 'value',
    'units', 'latitude', 'longitude', 'qualityCode',
    'remarkCode', 'depth','depthUnits', 'uncertainty',
    'collectMethod', 'weatherCode'
]
wq_clean = wq_data[col_to_keep].copy()
wq_clean.head() # clean dataset of raw data and relevant columns
```

```python
print(wq_clean.shape)
print(wq_clean.isnull().sum()) # NaN values in each column 
```

```python
print(f"Stations with coordinates: {wq_clean['latitude'].notna().sum()}")
print(f"Stations missing coordinates: {wq_clean['latitude'].isna().sum()}")
```

```python
wq_filter = wq_clean.pivot_table(
    index=["collectDate", "station", "latitude", "longitude"],
    columns="parameter",
    values="value",
    aggfunc="mean"
).reset_index()
wq_filter.head() #pivot table with parameters as columns 
```

```python
# including units in the pivot table causes issues,
# temporariily solved by including lookup table for units
units_lookup = wq_clean.groupby("parameter")["units"].agg(
    lambda x: x.value_counts().index[0]  
).reset_index()
units_lookup.columns = ["parameter", "units"]
units_lookup
```

```python
print(wq_filter.columns.tolist())
print(wq_filter.head(10))
print(wq_filter.shape)
```

```python
wq_filter["collectDate"] = pd.to_datetime(wq_filter["collectDate"])

wq_filter["date"] = wq_filter["collectDate"].dt.date

wq_daily = wq_filter.groupby(
    ["date", "station", "latitude", "longitude"]
)[[
    "CHLOROPHYLL-A(LC)",
    "SALINITY",
    "SECCHI DISK DEPTH",
    "TOTAL SUSPENDED SOLIDS",
    "TURBIDITY"
]].mean().reset_index()

wq_daily = wq_daily.rename(columns={
    "date": "Date",
    "station": "Station",
    "latitude": "Latitude",
    "longitude": "Longitude",
    "CHLOROPHYLL-A(LC)": "Chlorophyll-a",
    "SALINITY": "Salinity",
    "SECCHI DISK DEPTH": "Secchi Depth",
    "TOTAL SUSPENDED SOLIDS": "TSS",
    "TURBIDITY": "Turbidity"
})

print(f"Before aggregation: {wq_filter.shape[0]} rows")
print(f"After aggregation: {wq_daily.shape[0]} rows")
wq_daily.head()
```

```python
duplicates = wq_daily[wq_daily.duplicated(
    subset=["Date", "Station"], keep=False
)]

print(f"Rows with duplicate station/date combinations: {duplicates.shape[0]}")
```

```python
wq_daily.columns.name = None
wq_daily = wq_daily.set_index(["Date"])
wq_daily.head()
# Date is index with each parameter displaying as a column 
```
